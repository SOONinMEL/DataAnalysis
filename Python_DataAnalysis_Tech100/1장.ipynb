## 01 데이터를 읽어 들이자

```python
import pandas as pd
```

```python
# 고객데이터 (이름,성별...)
customer_master=pd.read_csv('customer_master.csv')
customer_master.head()
```

```python
item_master=pd.read_csv('item_master.csv')
item_master
```

```python
# 구매내역 데이터
transaction_1=pd.read_csv('transaction_1.csv')
transaction_1.head()
```

```python
# 구매내역 상세 데이터
transaction_detail_1=pd.read_csv('transaction_detail_1.csv')
transaction_detail_1.head()
```

```python
# 구체적인 목적이든 상관없이 먼저 데이터 전체를 파악하는 것이 중요
# 되도록 상세하게 나와 있는 쪽에 맞추어 데이터를 가공하는 것이 중요
```

## 02 데이터를 결합(union)해 보자

```python
# transaction2 데이터 불러들이기
transaction_2=pd.read_csv('transaction_2.csv')
transaction_2
```

```python
# transaction1 과 transaction2결합
transaction=pd.concat([transaction_1,transaction_2],ignore_index=True)
transaction
```

```python
# 제대로 결합되었는지 확인
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
# transaction_2_detail 데이터 불러들이기
transaction_detail_2=pd.read_csv('transaction_detail_2.csv')
transaction_detail_2.head()
```

```python
# transaction_detail_1 과 transaction_detail_2 결합
transaction_detail=pd.concat([transaction_detail_1,transaction_detail_2],ignore_index=True)
transaction_detail.head()
```

```python
# 이것도 제대로 결합되었는지 확인
print(len(transaction_detail_1))
print(len(transaction_detail_2))
print(len(transaction_detail))
```

## 03 매출 데이터끼리 결합(조인)해 보자

```python
# 데이터를 조인할 때는 기준이 데이터를 정확하게 결정하고, 어떤 칼럼을 키로 조인할지 생각한다.
# 1) 부족한(추가하고 싶은) 데이터 칼럼이 무엇인가, 2) 공통되는 데이터 칼럼은 무엇인가
```

```python
# 매출 데이터끼리 결합(조인) 해보자
# transaction_detail 데이터에 transaction의 'transaction_id','payment_date','customer_id'데이터 추가
join_data=pd.merge(transaction_detail,transaction[['transaction_id','payment_date','customer_id']],
                  on='transaction_id',how='left')
join_data.head()
```

```python
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

```python
# 조인할 데이터의 조인키에 중복 데이터가 존재하는 경우에는 데이터 개수가 늘어날 수 있기 때문에 주의해야한다.
```

## 04 마스터데이터를 결합(조인)해 보자

```python
# join data와 customer_master data, item_master join

join_data=pd.merge(join_data,customer_master,on='customer_id',how='left')
join_data=pd.merge(join_data,item_master,on='item_id',how='left')
join_data.head()
```

## 05 필요한 데이터 컬럼을 만들자

```python
# 매출컬럼 만들기
join_data['price']=join_data['quantity']*join_data['item_price']
join_data[['quantity','item_price','price']].head()
```

```python
# 데이터를 결합할 때는 신중히 갯수를 확인해야 한다.
# 되도록 검산이 가능한 데이터 컬럼을 찾고 계산해야한다.
```

## 06 데이터를 검산하자

```python
# 데이터 가공 전 transaction의 price 총합과 데이터 가공 후에 계산한 price의 총합비교하기
# 같은 값이 나오면 가공을 정확하게 한것
print(join_data['price'].sum())
print(transaction['price'].sum())
```

```python
# True/False로 검산하기
print(join_data['price'].sum()==transaction['price'].sum())
```

## 08 월별로 데이터를 집계해 보자

```python
# 전체 데이터를 한 번에 분석하면 데이터의 시계열 변화를 잘못 파악하는 경우가 있다.
# 이런 경우에는 데이터 범위를 좁혀서 분석하는 방법도 하나의 방법이다.
# 매출 데이터 같은 경우에는 매출이 늘어나고 있는지 줄어들고 있느지를 파악하는 것이 분석의 첫걸음이다.
```

```python
# data type check
join_data.dtypes
```

```python
# change payment_date's type to datetime
# dt-년,월 추출
# strftime-년,월 작성
join_data['payment_date']=pd.to_datetime(join_data['payment_date'])
join_data['payment_month']=join_data['payment_date'].dt.strftime('%Y%m')
join_data[['payment_date','payment_month']].head()
```

```python
# Monthly price
join_data.groupby(['payment_month'])['price'].sum()
```

## 09 월별, 상품별로 데이터를 집계해 보자

```python
# aggregating data by monthly and product
join_data.groupby(['payment_month','item_name'])[['price','quantity']].sum()
```

```python
pd.pivot_table(data=join_data,values=['price','quantity'],columns='payment_month',index='item_name',aggfunc='sum')
```

```python
# 표 형싱의 데이터는 세세한 숫자를 파악 할 수는 있지만, 수치가 늘고 잇는지, 감소하고 있는지 한눈에 파악하기 어렵다
# 분석의 목표는 어디까지나 현장에서 적절히 운용되고 시책을 마련하는 데 있다.
# 현장에는 숫자에 익수하지 못한 사람들도 있기 때문에 어떻게 전달하는가가 매우 중요
```

## 10 상품별 매출 추이를 가시화 해보자

```python
# 그래프용 데이터 pivot_table이용해서 만들기
graph_data=pd.pivot_table(join_data,index='payment_month',columns='item_name',values='price',aggfunc='sum')
graph_data.head()
```

```python
# 선그래프로 표시하기
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(list(graph_data.index),graph_data['PC-A'],label='PC-A')
plt.plot(list(graph_data.index),graph_data['PC-B'],label='PC-B')
plt.plot(list(graph_data.index),graph_data['PC-C'],label='PC-C')
plt.plot(list(graph_data.index),graph_data['PC-D'],label='PC-D')
plt.plot(list(graph_data.index),graph_data['PC-E'],label='PC-E')
plt.legend()
```
