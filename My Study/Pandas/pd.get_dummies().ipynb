```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'key':list('bbacab'),'data1':range(6)})
df
```

```python
pd.get_dummies(df['key'])
```

```python
dummies=pd.get_dummies(df['key'],prefix='key')
dummies
```

```python
df[['data1']].join(dummies) # merging the result with the orginal DataFrame
```

```python
values=np.random.randn(10)
values
```

```python
bins=[0,0.2,0.4,0.6,0.8,1] # discretization function
pd.get_dummies(pd.cut(values,bins))
```

```python
df=pd.DataFrame({'A':['a','b','a'],'B':['c','c','b'],'C':[1,2,3]})
df
```

```python
pd.get_dummies(df)
```

```python
pd.get_dummies(df,columns=['A']) # controlling the columns
```

```python
# As with the Series version, you can pass values for the prefix and prefix_sep. By default the column name is used as the prefix, and _ as the prefix separator
```

```python
simple=pd.get_dummies(df,prefix='new_prefix') # string: Use the same value for prefix or prefix_sep for each column to be encoded
simple
```

```python
from_list=pd.get_dummies(df,prefix=['from_A','from_B']) # Must be the same length as the number of columns being encoded
from_list
```

```python
from_dict=pd.get_dummies(df,prefix={'B':'from_B','A':'from_A'}) #Mapping column name to prefix
from_dict
```

```python
s=pd.Series(list('abcaa'))
s
```

```python
pd.get_dummies(s)
```

```python
pd.get_dummies(s,drop_first=True) #Sometimes it will be useful to only keep k-1 levels of a categorical variable to avoid collinearity when feeding the result to statistical models
```

```python
df=pd.DataFrame({'A':list('aaaaa'),'B':list('ababc')})
df
```

```python
pd.get_dummies(df)
```

```python
pd.get_dummies(df,drop_first=True) # When a column contains only one level, it will be omitted in the result
```

```python
# By default new columns will have np.uint8 dtype. To choose another dtype, use the dtype argument:

df=pd.DataFrame({'A':list('abc'),'B':list([1.1,2.2,3.3])})


pd.get_dummies(df,dtype=bool).dtypes
```
