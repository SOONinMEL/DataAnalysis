## 2. 함수나 매핑을 이용해서 데이터 변형하기

```python
data=pd.DataFrame({'food':['bacon','pulled pork','bacon',
                         'Pastrami','corned beef','Bacon',
                         'pastrami','honey ham','nova lox'],
                 'ounces':[4,3,12,6,7.5,8,3,5,6]})
data
```

```python
# 컬럼 추가(해당 육류가 어떤 동물의 고기인지 알려주는 칼럼)
meat_to_animal={
    'bacon':'pig',
    'pulled pork':'pig',
    'pastrami':'cow',
    'corned beef':'cow',
    'honey ham':'pig',
    'nova lox':'salmon'
}
```

```python
lowercased=data['food'].str.lower() # 대문자 데이터들을 소문자로 변경
```

```python
lowercased
```

```python
data['animal']=lowercased.map(meat_to_animal) # 사전류의 객체나 어떤 함수를 받는다.
data
```

```python
data['food'].map(lambda x:meat_to_animal[x.lower()]) # 함수이용.
```

```python
# map 메서드는 데이터의 요소별 변환 및 데이터를 다듬는 작업을 편리하게 수행할 수 있게 해준다.
```

## 3. 값 치환하기

```python
# fillna-일반적인 값 치환
# map- 한 객체 안에서 값의 부분집합 변경
# replace- 같은 작업에 대해 좀 더 간단하고 유연한 방법 제공
```

```python
data=pd.Series([1.,-999.,2.,-999.,-1000.,3.])
data
```

```python
data.replace(-999,np.nan) # NA값으로 치환
```

```python
# 여러개의 값을 한번에 치환
data.replace([-999,-1000],np.nan) #리스트로 넘겨준다.
```

```python
# 치환하려는 값마다 다른 값으로 치환
data.replace([-999,-1000],[np.nan,0]) #리스트로 넘겨준다.
```

```python
# 리스트대신 사전이용
data.replace({-999:np.nan,-1000:0})
```

## 4.축 색인 이름 바꾸기

```python
data=pd.DataFrame(np.arange(12).reshape((3,4)),
                 index=['Ohio','Colorado','New York'],
                 columns=['one','two','three','four'])
data
```

```python
# map 메서드와 lambda함수 이용해서 축 색인하기
transform=lambda x:x[:4].upper()
data.index.map(transform)
```

```python
# 대문자로 변경된축 바로 DataFrame Index에 적용하기
data.index=data.index.map(transform)
data
```

```python
# 원래 객체를 변경하지 않고 새로운 객체 생성하려면 rename메서드 사용
data.rename(index=str.title,columns=str.upper)
```

```python
# rename메서드는 사전형식의 객체를 이용해서 축 이름 중 일부만 변경하는 것도 가능
data.rename(index={'OHIO':'INDIANA'},
           columns={'three':'peekaboo'})
```

```python
# 원본데이터를 바로 변경하려면 inplace=True옵션을 넘겨준다.
data.rename(index={'OHIO':'INDIANA'},
           columns={'three':'peekaboo'},
           inplace=True)
data
```

## 5.개별화와 양자화

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32] # 나이 데이터
```

```python
bins=[18,25,35,60,100] #18-25,26-35,35-60,60이상 그롭으로 나누기
cats=pd.cut(ages,bins)
cats
```

```python
cats.codes
```

```python
cats.categories #카테고리 객체
```

```python
pd.value_counts(cats) # pandas.cut결과에 대한 그룹 수
# 중괄호쪽 값은 포함하지 않고 대괄호쪽 값은 포함하는 간격
```

```python
pd.cut(ages,bins,right=False) #중괄호 대신 대괄호쪽이 포함되지 않도록 바꿔준다.
```

```python
# labels 옵션을 통해 그룹의 이름을 직접 넘겨줄 수도 있다.
group_names=['Youth','YoungAdult','MiddleAged','Senior']
pd.cut(ages,bins,labels=group_names)
```

```python
# 그룹의 경곗값을 넘기는 대신 그룹의 갯수를 넘겨준다.
# 데이터에서 최솟값과 최대값을 기준으로 균등한 길이의 그룹을 자동으로 생각한다.
data=np.random.rand(20)
pd.cut(data,4,precision=2)# 소수점 아래 2자리까지로 제한
```

```python
# qcut함수를 이용하면 표본 변위치를 기반으로 데이터를 나눠준다.
# cut함수는 데이터 분산에 따라 각각의 그룹마다 데이터 수가 다르게 나뉘어지는 경우가 많은 반면에
# qcut함수는 표준 변위치를 사용하기 때문에 적당히 같은 크기의 그룹으로 나눌 수 있다.
data=np.random.randn(1000) #정규분포
cats=pd.qcut(data,4)
```

```python
pd.value_counts(cats)
```

```python
# 변위치 직접 지정 가능 (변위치는 0부터1까지)

pd.qcut(data,[0,0.1,0.5,0.9,1.])
```

## 6. 특잇값(outlier)을 찾고 제외하기

```python
data=pd.DataFrame(np.random.randn(1000,4))
data.describe()
```

```python
# 한컬럼에서 절댓값이 3을 초과하는 값을 찾아내기
col=data[2]
col[np.abs(col)>3]
```

```python
# 절댓값이 3을 초과하는 값이 들어있는 모든 로우를 선택
# DataFrame에서 any메서드를 사용한다.
data[(np.abs(data)>3).any(1)]
```

```python
# -3이나 3을 초과하는 값을 -3또는 3으로 지정
data[np.abs(data)>3]=np.sign(data)*3
data.describe()
```

```python
# np.sign(data) data값이 양수인지 음수인지에 따라 1이나 -1이 담긴 배열을 반환
np.sign(data).head()
```

## 8.표시자/더미 변수 계산하기

* 통계 모델이나 머신러닝 에플리케이션을 위한 데이터변환 방법 (분류값을 '더미'나 '표시자' 행렬로 전환시킨다.)
* 한 컬럼에 k가지의 값이 있다면 k개의 컬럼이 있는 df나 행렬을 만들고 1과0을 채워 넣는것

```python
df=pd.DataFrame({'key':['b','b','a','c','a','b'],
                'data1':range(6)})
df
```

```python
pd.get_dummies(df['key'])
```

```python
# prefix(접두어)를 추가하여 다른 데이터와 병합
dummies=pd.get_dummies(df['key'],prefix='key')
df_with_dummy=df[['data1']].join(dummies)
df_with_dummy
```

```python
mnames=['move_id','title','genres']
movies=pd.read_table('movies.dat',sep='::',header=None,names=mnames)
movies[:10]
```

```python
# 장르 출력
all_genres=[]
for x in movies['genres']:
    all_genres.extend(x.split('|'))
genres=pd.unique(all_genres)
genres
```

```python
# 표시자 데이터프레임 생성(0으로 초기화된 데이터프레임 생성)
zero_matrix=np.zeros((len(movies),len(genres)))
dummies=pd.DataFrame(zero_matrix,columns=genres)
dummies
```

```python
# 각 장르의 컬럼색인을 계산(dummies.columns사용)
gen=movies.genres[0]
gen.split('|')
```

```python
dummies.columns.get_indexer(gen.split('|'))
```

```python
# 각 영화를 순회하면서 dummies의 각 로우의 항목을 1로 설정
for i,gen in enumerate(movies.genres):
    indices=dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i,indices]=1
```

```python
dummies
```

```python
# movies와 dummies조합
movies_windic=movies.join(dummies.add_prefix('Genre_'))
movies_windic
```

```python
movies_windic.iloc[0]
```
