```python
import pandas as pd
import numpy as np
```

## 1. 데이터베이스 스타일로 DataFrame합치기

```python
# 하나 이상의 키를 사용해서 뎅터 집합의 로우를 합친다.
# pd.merge()함수를 이용해서 알고리즘을 데이터에 적용시킨다.
df1=pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                 'data1':range(7)})

df2=pd.DataFrame({'key':['a','b','d'],
                 'data2':range(3)})
```

```python
df1
```

```python
df2
```

```python
# 다대일인 경우
# 어떤 컬럼을 병합할 것인지 명시하지 않을 겨우, merge함수는 중복된 컬럼 이름을 key로 사용한다.
pd.merge(df1,df2)
```

```python
# 명시적으로 key를 지정해주자.
pd.merge(df1,df2,on='key')
```

```python
df3=pd.DataFrame({'1key':['b','b','a','c','a','a','b'],
                 'data1':range(7)})
df4=pd.DataFrame({'rkey':['a','b','d'],
                 'data2':range(3)})
```

```python
df3
```

```python
df4
```

```python
# 중복된 컬럼이 없을 경우에는 따로 지정가능!
pd.merge(df3,df4,left_on='1key',right_on='rkey')
```

```python
# 다대다 병합
df1=pd.DataFrame({'key':['b','b','a','c','a','b'],
                 'data1':range(6)})

df2=pd.DataFrame({'key':['a','b','a','b','d'],
                 'data2':range(5)})
```

```python
df1
```

```python
df2
```

```python
# inner join (default)- 양쪽 테이블에 존재하는 키 조합을 사용한다.
pd.merge(df1,df2,on='key',how='inner')
```

```python
# left join- 왼쪽테이블에 존재하는 모든 키 조합을 사용한다.
pd.merge(df1,df2,on='key',how='left')
```

```python
# right join-오른쪽테이블에 존재하는 모든 키 조합을 사용한다.
pd.merge(df1,df2,on='key',how='right')
```

```python
# outer join- 양쪽 테이블에 존재하는 모든 키 조합을 사용한다.
pd.merge(df1,df2,on='key',how='outer')
```

```python
# 여러 개의 키를 병합하기
left=pd.DataFrame({'key1':['foo','foo','bar'],
                  'key2':['one','two','one'],
                  'lval':[1,2,3]})

right=pd.DataFrame({'key1':['foo','foo','bar','bar'],
                   'key2':['one','one','one','two'],
                   'rval':[4,5,6,7]})
```

```python
left
```

```python
right
```

```python
# 컬럼 이름이 담긴 리스트를 넘기면 여러 개의 키를 병합 할 수있다.
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
# 겹치는 컬럼 이름에 대한 처리
pd.merge(left,right,on='key1')
```

```python
pd.merge(left,right,on='key1',suffixes=('_left','_right')) # 컬럼 이름 뒤에 붙일 문자열을 지정해준다.
```

## 2.색인 병합하기
* 병합하려는 키가 DataFrame의 색인일 경우가 있다.
* 이런경우에는 left_index=True or right_index=Ture 을 지정해서 해당 색인을 병합키로 사용할 수 있다.

```python
left1=pd.DataFrame({'key':['a','b','a','a','b','c'],
                   'value':range(6)})
right1=pd.DataFrame({'group_val':[3.5,7]},index=['a','b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)
```

```python
# 외부 조인을 실행해서 합집합을 구한다.
pd.merge(left1,right1,left_on='key',right_index=True,how='outer')
```

* 계층된 색인 데이터

```python
lefth=pd.DataFrame({'key1':['Ohio','Ohio','Ohio','Nevada','Nevada'],
                   'key2':[2000,2001,2002,2001,2002],
                   'data':np.arange(5.)})
righth=pd.DataFrame(np.arange(12).reshape((6,2)),
                   index=[['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],
                         [2001,2000,2000,2000,2001,2002]],
                   columns=['event1','event2'])
```

```python
lefth
```

```python
righth
```

```python
# 계층 색인된 데이터를 병합할때에는 리스트로 여러 개의 컬럼을 지정해서 병합해야한다.
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True)
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True,how='outer') # 중복되는 색인값을 다룰 때는 how='outer' 옵션을 사용한다.
```

```python
left2=pd.DataFrame([[1.,2.],[3.,4.],[5.,6.]],
                  index=['a','c','e'],
                  columns=['Ohio','Nevada'])

right2=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[13,14]],
                   index=['b','c','d','e'],
                   columns=['Missouri','Alabama'])
```

```python
left2
```

```python
right2
```

```python
# 양쪽에 공통적으로 존재하는 여러 개의 색인을 병합

pd.merge(left2,right2,how='outer',left_index=True,right_index=True)
```

```python
# 색인으로 병합할 때 DataFrame의 join 메서드를 사용하면 편리하다.
# join 메서드는 컬럼이 겹치지 않으며 완전히 같거나 유사한 색인 구조를 가진 여러 개의 DataFrame 객체를 병합할 때 사용할 수 있다.
left2.join(right2,how='outer')
```

```python
# join 메서드를 호출한 후, DataFrame의 컬럼 중 하나에 대해 조인을 수행하기
left1
```

```python
right1
```

```python
left1.join(right1,on='key')
```

```python
# 색인 대 색인으로 두 DataFrame병합하기
```

```python
another=pd.DataFrame([[7.,8.,],[9.,10,],[11.,12.],[16.,17,]],
                    index=['a','c','e','f'],
                    columns=['New York','Oregon'])
another
```

```python
left2.join([right2,another])
```

```python
left2.join([right2,another],how='outer')
```

## 3.축 따라 이어붙이기
* 이어붙이기 (concatenation)- numpy에서 제공하는 함수

```python
arr=np.arange(12).reshape((3,4))
arr
```

```python
np.concatenate([arr,arr],axis=1)
```

```python
# 색인이 겹치지 않는 3개의 Series
s1=pd.Series([0,1],index=['a','b'])
s2=pd.Series([2,3,4],index=['c','d','e'])
s3=pd.Series([5,6],index=['f','g'])
```

```python
# 객체를 리스트로 묶어서 concat함수에 전달한다
pd.concat([s1,s2,s3])
```

```python
# axis=1을 넘겼을때
pd.concat([s1,s2,s3],axis=1)
```

```python
s4=pd.concat([s1,s3])
s4
```

```python
pd.concat([s1,s4],axis=1) #join='inner'를 넘겨서 교집합을 구한다.
```

```python
pd.concat([s1,s4],axis=1,join='inner') # inner옵션으로 인해 'f' 와 'g' 라벨이 사라진다.
```

```python
# 이어붙인 축에 대해 계층적 색인을 생성하여 식별이 가능하도록 하게 해준다.
result=pd.concat([s1,s2,s3],keys=['one','two','three'])
result
```

```python
result.unstack()
```

```python
pd.concat([s1,s2,s3],axis=1,keys=['one','two','three']) # axis=1로 병합할 경우 keys DataFrame의 컬럼 제목이 된다.
```

### DataFrame객체에 concat 방식 적용하기

```python
df1=pd.DataFrame(np.arange(6).reshape(3,2),index=['a','b','c'],
                columns=['one','two'])
df1
```

```python
df2=pd.DataFrame(5+np.arange(4).reshape(2,2),index=['a','c'],
                columns=['three','four'])

df2
```

```python
pd.concat([df1,df2],axis=1,keys=['level1','level2'])
```

```python
# 리스트 대신 객체의 사전을 넘기면 사전의 키가 keys 옵션으로 사용된다.
pd.concat({'level1':df1,'level2':df2},axis=1)
```

```python
pd.concat([df1,df2],axis=1,keys=['level1','level2'],names=['upper','lower']) # 새로 생성된 계층의 이름 지정
```

```python
df1=pd.DataFrame(np.random.randn(3,4),columns=['a','b','c','d'])
df1
```

```python
df2=pd.DataFrame(np.random.randn(2,3),columns=['b','d','a'])
df2
```

```python
pd.concat([df1,df2])
```

```python
# DataFrame의 로우 색인이 분석에 필요한 데이터를 포함하고 있지 않은 경우 ignore_index=True옵션을 주어준다.
# 이어붙인 축의 색인을 유지하지 않고 range(total_length)로 새로운 색인을 생성한다.
pd.concat([df1,df2],ignore_index=True)
```
