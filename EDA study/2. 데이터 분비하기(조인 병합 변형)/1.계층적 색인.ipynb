```python
import pandas as pd
import numpy as np
```

```python
# Multiindex를 색인으로 하는 Series
# 색인의 계층을 보여준다.
data=pd.Series(np.random.randn(9),
              index=[['a','a','a','b','b','c','c','d','d'],
                    [1,2,3,1,3,1,2,2,3]])
data
```

```python
# partial indexing(부분적 색인으로 접근)
data['b']
```

```python
data['b':'c']
```

```python
data.loc[['b','d']]
```

```python
# 하위 계층의 객체를 선택
data.loc[:,2]
```

```python
# unstack 메서드를 사용해서 데이터를 새롭게 배열하기
data.unstack()
```

```python
# stack: unstack의 반대작업
data.unstack().stack()
```

```python
# 인덱스와 컬럼 모두 계층적 색인을 가질때
frame=pd.DataFrame(np.arange(12).reshape((4,3)),
                 index=[['a','a','b','b,'],[1,2,1,2]],
                 columns=[['Ohio','Ohio','Colorado'],
                         ['Green','Red','Green']])
frame
```

```python
# 계층적 색인의 각 단계에 이름가지게 하기
frame.index.names=['key1','key2']
frame.columns.names=['State','color']
frame
```

```python
frame['Ohio']
```

## 1. 계층의 순서를 바꾸고 정렬하기

```python
# swaplevel: 넘겨받은 두 개의 계층 번호나 이름이 뒤바뀐 새로운 객체를 반환한다.(데이터는 변경되지 않는다.)
frame
```

```python
frame.swaplevel('key1','key2')
```

```python
frame.sort_index(level=1)
```

```python
frame.swaplevel(0,1).sort_index(level=0)
```

## 2.계층별 요약 통계

```python
frame
```

```python
# 기술 통계와 요약 통계는 level 옵션을 가지고 있다.
# 어떤 한 축에 대해 합을 구하고 싶은 단계를 지정할 수 있는 옵션
frame.sum(level='key2')
```

```python
frame.sum(level='color',axis=1)  # 컬럼의 단계를 지정하여 컬럼 한축에 대한 값을 구한다.
```

## 3. DataFrame의 컬럼 사용하기

```python
frame=pd.DataFrame({'a':range(7),'b':range(7,0,-1),
                   'c':['one','one','one','two','two','two','two'],
                   'd':[0,1,2,0,1,2,3]})
frame
```

```python
# set_index: 하나 이상의 컬럼을 색인으로 만들어준다.
frame2=frame.set_index(['c','d'])
frame2
```

```python
# drop=False를 사용하여 컬럼을 명시적으로 남겨둔 다음에 DataFrame에서 삭제되지 않게 한다.
frame.set_index(['c','d'],drop=False)
```

```python
# 계층적 색인 단계가 컬럼으로 이동한다.
frame2.reset_index()
```
