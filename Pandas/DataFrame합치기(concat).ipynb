```python
import pandas as pd
```

```python
# df1 customer 정보

df1=pd.DataFrame({'id':[1,2,3],
                 'customer_id':[1,2,3],
                 'customer_name':['James','Lucas','Jason']})
df1
```

```python
# df2 order정보

df2=pd.DataFrame({'id':[1,2,4],
                 'order_id':[120,121,122],
                 'order_location':['Seoul','Busan','Daegu']})

df2
```

```python
# concat(): 두 데이터프레임을 연결해서 하나의 데이터프레임으로 만든다.
# 두 데이터 프레이을 위/아래 또는 왼쪽/오른쪽으로 연결하기만 함
# pd.concat([DataFrame1,DataFrame2])
```

```python
# axis=0을 기본값으로 하여 새로운 Series객체를 생성
concat1=pd.concat([df1,df2])
concat1
```

```python
# axis=1이면 DataFrame객체를 형성
concat2=pd.concat([df1,df2],axis=1)
concat2
```

```python
# join='outer'와 join='inner'
# df3 3반의 시험 성적표

df3=pd.DataFrame({'Math':[80,90,95],'Science':[75,70,60],'English':[85,65,90]},index=['A','B','C'])
df3
```

```python
# df4 4반의 시험 성적표

df4=pd.DataFrame({'Math':[70,84,92],'Korean':[88,75,65],'English':[75,86,92]},index=['A','B','D'])
df4
```

```python
# join='outer' default 값

concat4=pd.concat([df3,df4],join='outer')
concat4
```

```python
concat5=pd.concat([df3,df4],axis=1,join='outer')
concat5
```

```python
# column Korean하고 Science는 join='inner' 옵션으로 인해 사라졌다.
concat6=pd.concat([df3,df4],join='inner')
concat6
```

```python
# index C와 D는 join='inner' 옵션으로 인해 사라졌다.
concat7=pd.concat([df3,df4],axis=1,join='inner')
concat7
```

```python
# verifiy integrity
# 이어붙인 객체에 중복되는 축이 있는지 검사하고 있다면 예외를 발생시킨다.

concat7=pd.concat([df3,df4],verify_integrity=True)
concat7
```

```python
concat8=pd.concat([df3,df4],verify_integrity=False)
concat8
```

```python
# ignore_index=True:
# 이어붙인 축의 색인을 유지하지 않고 range(total_length)로 색인을 생성한다.
# DataFrame의 로우 색인이 분석에 필요한 데이터를 포함하고 있지않는 경우 유용하다.

concat9=pd.concat([df3,df4],ignore_index=True)
concat9
```

```python
# range(total_length)로 컬럼생성
concat10=pd.concat([df3,df4],axis=1,ignore_index=True)
concat10
```

```python
# keys
# 이어붙일 객체나 이어붙인 축에 대한 계층 색인을 생성해주는 기능
# Series를 이어붙이기 전의 개별Series를 구분할 수 있게 해준다.

concat11=pd.concat([df3,df4],keys=['A1','B1'])
concat11
```

```python
# axis=1로 병합할경우 column의 제목이된다.
concat12=pd.concat([df3,df4],axis=1,keys=['A1','B1'])
concat12
```
