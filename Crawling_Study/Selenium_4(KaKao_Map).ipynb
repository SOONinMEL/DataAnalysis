```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import time
```

```python
driver=wb.Chrome()
driver.get('https://map.kakao.com/')
```

```python
#1. 검색창 위치알아오기
# #은 id 구분자 .은 클래스 구분자, .을 클래스 구분자로 구분하는 것을 방지하기 위해 \을 통해 .을 문자처리하게 인식하게 해준다.
search=driver.find_element_by_css_selector('#search\.keyword\.query')
```

```python
img=driver.find_element_by_css_selector('div.inner_coach_layer')
img.click()
```

```python
#2.키워드를 입력
search.send_keys('광주맛집')
```

```python
#3. 검색버튼 실행
searchBtn=driver.find_element_by_css_selector('#search\.keyword\.submit')
searchBtn.click()
```

```python
# 장소 더보기 버튼 클릭
btnMore=driver.find_element_by_css_selector('#info\.search\.place\.more')
btnMore.click()
```

```python
# 페이지번호 데이터 
# 5개의 page가 들어있는 변수
page=driver.find_elements_by_css_selector('#info\.search\.page>div>a')
```

```python
try:
    while True:
        for i in range(6):
            if (i<5):
                page[i].click()
                time.sleep(2)
                soup=bs(driver.page_source,'lxml')
                ad=soup.select('span.ico_ad')
                # 불필요한 요소 추출하기(extract())
                # BeautifulSoup를 이용하는 이유(extract라는 기능을 사용하기 위해)
                for i in ad:
                    i.extract()
                title=soup.select('a.link_name')
                for i in title:
                    print(i.text)
            else:
                more=driver.find_element_by_css_selector('#info\.search\.page\.next')
                more.click()
except:
    print('Finish collecting data')
```

```python
driver.quit()
```
